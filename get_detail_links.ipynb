```python
from bs4 import BeautifulSoup
import requests

!pip3 install beautifulsoup4

url = 'https://lalafo.kg/osh/zemelnye-uchastki/prodazha-zemli'

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    a_tags = soup.find_all('a')
    p_tags = soup.find_all('p')

    for a_tag, p_tag in zip(a_tags, p_tags):
        print(f"title: {a_tag.get_text()}, price: {p_tag.get_text()}")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")

```

```python
from bs4 import BeautifulSoup
import requests

# Install BeautifulSoup if not already installed
!pip3 install beautifulsoup4

# Define the URL of the web page you want to scrape
url = 'https://lalafo.kg/osh/zemelnye-uchastki/prodazha-zemli'

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    a_tags = soup.find_all('a')
    p_tags = soup.find_all('p')
    data_links = []

    for a_tag, p_tag in zip(a_tags, p_tags):
        a_href = a_tag.get('href')
        if 'user' not in a_href:
          data_links.append('https://lalafo.kg'+a_href)
          # print(f"identical pages link: {a_href}")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")


print(data_links)
```
